```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/kaggle/input/ccdata/CC GENERAL.csv')
```

```python
plt.scatter(data['BALANCE'],data['PAYMENTS'])
plt.xlabel('Balance')
plt.ylabel('Payments')
plt.show()
```

```python
my_data = data.loc[:,['BALANCE','PAYMENTS']]

from sklearn.cluster import KMeans
inertias = []
for k in range(1,10):
    kmeans = KMeans(n_clusters = k,n_init='auto',init='k-means++')
    kmeans.fit(my_data)
    inertias.append(kmeans.inertia_)
plt.plot(range(1,10),inertias,'-o')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show
```

```python
kmeans=KMeans(n_clusters=4, init='k-means++', n_init='auto')
kmeans.fit(my_data)
labels = kmeans.predict(my_data)
plt.scatter(data['PAYMENTS'],data['BALANCE'],c = labels)
plt.xlabel('PAYMENTS')
plt.xlabel('BALANCE')
plt.show()
```
